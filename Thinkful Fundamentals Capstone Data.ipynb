```python
N = 3
w_ed_means = (1.052397, 3.456456, 2.988258)
h_ed_means = (0.902869, 3.663664, 3.459883)   
w_std = (1.052397, 0.796488, 0.966020)
h_std = (0.902869, 0.707810, 0.728856)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, w_ed_means, width, yerr=w_std)
p2 = plt.bar(ind, h_ed_means, width,
             bottom=h_ed_means, yerr=h_std)

plt.ylabel('Mean Scores')
plt.title('Mean Scores by Group and Gender')
plt.xticks(ind, ('No Contraception', 'Long-term Use', 'Short-term Use'))
plt.yticks(np.arange(0, 8.5, 1))
plt.legend((p1[0], p2[0]), ('Women', 'Men'))

plt.show()
```
